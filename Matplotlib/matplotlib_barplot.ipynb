```python
from matplotlib import pyplot as plt
import numpy as np
```

# Bar plot

A bar plot with numerical values on the x axis, y axis random with errorbars.

```python
# Create the data to plot
x = np.arange(8)
y = np.random.randint(25, 50, 8)
yerrs = np.sqrt(y)
yerrs
```

```python
# Create the plot
plt.bar(x,y, yerr=yerrs)
```

## Categorical Data

x axis values will often be categorical - strain, mutant, etc. Create a list of strings for the x axis labels. In this example, we will plot soy and corn on two seperate plots.

```python
names = ['Missouri', 'Kansas',  'Iowa',  'Nebraska', 'Ohio', 'Minnesota', 'Illinois']
corn = [466, 645, 2508, 1787, 617, 1360, 2278]

plt.bar(names, corn)
plt.xticks(rotation=45)
plt.title("Corn production in some states")
plt.xlabel("State")
plt.ylabel("Million Bushels")
```

```python
soy = [230, 186, 501, 283, 209, 297, 532]
plt.bar(names, soy)
plt.xticks(rotation=45)
plt.title("Soybean production in some states")
plt.xlabel("State")
plt.ylabel("Million Bushels")
```

## Grouped Bar Plots

We can also plot both soy and corn on the same plot by calling `plt.bar` twice, once with corn and once with soy.


```python
width = .4
xf = np.arange(len(names))
xcorn = xf - width/2
xsoy = xf + width/2

plt.bar(xcorn, corn, width=width, tick_label=names)
plt.bar(xsoy, soy, width=width)
plt.xticks(rotation=45, ticks=xf)
plt.title("Corn production in some states")
plt.xlabel("State")
plt.ylabel("Million Bushels")
```

# Other examples

```python
loc = np.random.randint(10,20, (10,))
scale = 3*np.random.rand(10)

# create data array that is 20 rows by 10 columns
data = np.random.normal(loc, scale, size=(400, 10))
```

```python
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
xd = np.arange(len(loc))
yd = data.mean(axis=0)
yerr = data.std(axis=0)/np.sqrt(data.shape[0])
plt.bar(xd, yd, yerr=yerr, tick_label=labels)
```

```python
bp = plt.boxplot(data, labels=labels)
```
