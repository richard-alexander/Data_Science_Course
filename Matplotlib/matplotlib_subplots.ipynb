```python
from matplotlib import pyplot as plt
import numpy as np
import tifffile
```

```python
image = tifffile.imread('mitosis.tif')
image.shape
```

```python
plt.imshow(image[25, 2, 0, :,:], cmap='Reds_r')
```

```python
xred = image[25, 2, 0, :,:]
xgreen = image[25, 2, 1, :,:]
hr, hrx = np.histogram(xred, bins=256)
hg, hgx = np.histogram(xgreen, bins=256)
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.imshow(xred, cmap='Reds_r')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(xgreen, cmap='Greens_r')
plt.axis('off')

plt.subplot(2, 1, 2)
plt.plot(hrx[:-1], hr, color='red')
plt.plot(hgx[:-1], hg, color='green')
plt.title("Histogram")
plt.yscale("log")
```
