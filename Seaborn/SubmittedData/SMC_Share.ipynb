```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/richard-alexander/Data_Science_Course/master/Seaborn/SubmittedData/SMC_LabTeamPublications.csv')
```

```python
df.pivot_table(index='Lab', aggfunc={'LastAuthor':len}, columns='Team').fillna(
    0)['LastAuthor'].plot.barh(figsize=[15,15], stacked=True)
```
